```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GOOGLE_API_KEY"] = os.getenv("GEMINI_API_KEY")
```

```python
# CALLING THE AI AGENT HERE
from langchain.agents import create_agent

def get_SKY(SKY: str) -> str:
    """What is Todays Sky."""
    return f"Todays {SKY} is clear"

agent = create_agent(
    model='google_genai:gemini-3.5-flash',
    tools=[get_SKY],
    system_prompt="You are a helpful assistant"
)
# agent # Shows The Agent WorkFlow
```

```python
response = agent.invoke({"messages": [{"role": "user", "content": "What is the sky today?"}]})
# print("response", response["messages"][-1])
response["messages"]
# response["messages"][-1].text # To Get Only The Answer From The Response Depending On The Model
```
